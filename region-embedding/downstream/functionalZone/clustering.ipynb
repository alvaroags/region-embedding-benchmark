```python
import pandas as pd
import numpy as np

df = pd.read_csv('/home/gegen07/dev/projects/region-embedding-benchmark/region-embedding/baselines/HGI/data/region_embedding.csv')
cta = pd.read_csv('../../data/cta_nyc.csv')
```

```python
df = df.merge(cta, on='BoroCT2020')
```

```python
df
```

```python
X = df.iloc[:, 0:64].values
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for x in range(2, 20):
    kmeans = KMeans(n_clusters=x, random_state=0, n_init="auto")
    print(f"Silhueta k={x}, silhueta={silhouette_score(X, kmeans.fit_predict(X))}")
```

```python
kmeans = KMeans(n_clusters=11, random_state=0, n_init="auto")
```

```python
df['labels'] = kmeans.fit_predict(X)
```

```python
df.to_csv('HGI_clustered.csv', index=False)
```
